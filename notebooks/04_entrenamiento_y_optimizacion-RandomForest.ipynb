```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
import joblib
```

```python
# Cargamos los datos procesados.
X_train = pd.read_csv("../data/2. processed/X_train.csv")
X_test = pd.read_csv("../data/2. processed/X_test.csv")
y_train = pd.read_csv("../data/2. processed/y_train.csv")
y_test = pd.read_csv("../data/2. processed/y_test.csv")
```

```python
rfc_base=RandomForestClassifier( n_estimators=100, random_state=42, min_samples_split=30, min_samples_leaf=100)
```

```python
rfc_base.fit(X_train,y_train)
```

```python
train_pred = rfc_base.predict(X_train) #Predicciones del modelo para Train
test_pred = rfc_base.predict(X_test)#Predicciones del modelo para test
```

```python
print("RANDOMFOREST")
print("Train")
print(classification_report(y_train, train_pred))
print("--------------------------------------------------")
print("Test")
print(classification_report(y_test, test_pred))
```

```python
#Guardamos lo predicho por el modelo.
joblib.dump(rfc_base, "../models/smoking_model_forest.pkl")
```

Optimización


```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid_rfc = {
    "n_estimators": [100],
    "max_depth": [10, 15, None],
    "class_weight": [None, "balanced"]
}
```

```python
grid_rfc = GridSearchCV(
    estimator=rfc_base,
    param_grid=param_grid_rfc,
    scoring="f1",   
    cv=5,
    n_jobs=4,
    verbose=0
)
```

```python
grid_rfc.fit(X_train, y_train)
```

```python
 best_rfc = grid_rfc.best_estimator_ # Con esto extraemos los "mejores" hiperparámetros encontrados por GridSearchCV
```

```python
best_rfc.fit(X_train, y_train) #Entrenamos el modelo
```

```python
train_pred_rfc = best_rfc.predict(X_train)
test_pred_rfc = best_rfc.predict(X_test)
```

```python
print("RANDOMFOREST best")
print("Train (best)")
print(classification_report(y_train, train_pred_rfc))
print("--------------------------------")
print("Test (best)")
print(classification_report(y_test, test_pred_rfc))
```

```python
#Guardamos el "mejor" modelo entrenado.
joblib.dump(best_rfc, "../models/smoking_model_forest_best.pkl")
```
