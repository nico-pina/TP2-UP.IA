```python
import pandas as pd
import joblib
```

```python
df_predic= pd.read_csv("../data/1. raw/smoking_prediction_entrega.xlsx - smoking_prediction.csv ")
```

```python
df_predic.head()
```

```python
X_predic = df_predic.drop(columns=["ID","oral"]) 
```

```python
X_predic
```

```python
#Cargamos los datos preprocesados.
ohe = joblib.load("../data/4. objetos entrenados/ohe.pkl")
scaler = joblib.load("../data/4. objetos entrenados/scaler.pkl")
num_cols = joblib.load("../data/4. objetos entrenados/num_cols.pkl")
cat_cols = joblib.load("../data/4. objetos entrenados/cat_cols.pkl")
feature_columns = joblib.load("../data/4. objetos entrenados/feature_columns.pkl")
```

```python
X_predic_ohe=ohe.transform(X_predic[cat_cols]) # Aplicamos OHE a lasa columnas categóricas de X_predic
```

```python
ohe_columns = ohe.get_feature_names_out(cat_cols) # Obtemos los nombres de las columnas creadas por el OneHotEncoder
```

```python
X_predic_ohe_df = pd.DataFrame(X_predic_ohe, columns=ohe_columns, index=X_predic.index)
```

```python
X_predic_num = X_predic[num_cols] #Seleccionamos solo las columnas numéricas de X_predic.
```

```python
X_predic_final = pd.concat([X_predic_num, X_predic_ohe_df], axis=1) # Concatenamos las columnas numéricas con las columnas OHE en un solo DataFrame final
```

```python
X_predic_final_scaled = X_predic_final.copy() # Generamos una copia de X_predic_final
```

```python
X_predic_final_scaled[num_cols] = scaler.transform(X_predic_final[num_cols]) #escalamos las variables numéricas
```

```python
X_predic_final_scaled = X_predic_final_scaled[feature_columns] #aseguramos mismo orden de columnas que en el entrenamiento
```

```python
X_predic_final_scaled.head()
```

```python
model_rf = joblib.load("../models/smoking_model_forest_best.pkl") 
```

```python
model_xgb = joblib.load("../models/smoking_model_xgb_best.pkl")
```

Predicciones de RandomForest

```python
predic_rf=model_rf.predict(X_predic_final_scaled) # Generamos la predicción del modelo RF con los datos escalados.
```

```python
resultado_rf=pd.DataFrame({"ID":df_predic["ID"],"smoking_pred_rf":predic_rf}) #Creamos un DataFrame con el ID y la predicción
```

```python
resultado_rf.head()
```

```python
entrega_rf=resultado_rf[["ID","smoking_pred_rf"]]
```

```python
entrega_rf.to_csv("../predictions_csv/smoking_predictions_rf.csv",index=False) #Exportamos la predicciones en un archivo .csv
```

Predicciones XGboost

```python
 predic_xgb=model_xgb.predict(X_predic_final_scaled) # Generamos la predicción del modelo XGB con los datos escalados.
```

```python
resultado_xgb=pd.DataFrame({"ID":df_predic["ID"], "smoking_pred_xgb":predic_xgb}) #Creamos un DataFrame con el ID y la predicción
```

```python
resultado_xgb.head()
```

```python
entrega_xgb=resultado_xgb[["ID","smoking_pred_xgb"]]  #Creamos un DataFrame con el ID y la predicción
```

```python
entrega_xgb.to_csv("../predictions_csv/smoking_predictions_xgb.csv",index=False)
```

Coincidencia entre modelos

```python
#unimos las dos  predicciones en un solo DF
compare = resultado_rf.merge(resultado_xgb, on="ID")
```

```python
compare["coinciden"] = (compare["smoking_pred_rf"] == compare["smoking_pred_xgb"]) # Creamos una columna booleana indicando si ambas predicciones coinciden
```

```python
compare["coinciden"].value_counts() #Contamos las coincidencias.
```

```python
porcentaje_coincidencia=compare["coinciden"].mean() #Calculamos el porcentaje de coicidencias.
```

```python
print(f"Los dos modelos coinciden en el {porcentaje_coincidencia:.2%} de los casos")
```
