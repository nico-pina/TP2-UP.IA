```python
import pandas as pd 
import matplotlib.pyplot as plt
import joblib
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
model_xgb = joblib.load("../models/smoking_model_xgb_best.pkl")
X_test = pd.read_csv("../data/2. processed/X_test.csv")
y_test = pd.read_csv("../data/2. processed/y_test.csv")
```

```python
y_pred_xgb = model_xgb.predict(X_test)
```

```python
cm = confusion_matrix(y_test, y_pred_xgb) # Calculamos la matriz de confusión.
```

```python
#Graficamos la amtriz de confusión.
fig, ax = plt.subplots(figsize=(6, 5))
display = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["No Fumador (0)", "Fumador (1)"])
display.plot(cmap="Blues", ax=ax, values_format="d") 

plt.title("Matriz de Confusión - XGBoost Optimizado", fontsize=14, pad=15)
plt.xlabel("Predicción del Modelo", fontsize=12)
plt.ylabel("Clase Real", fontsize=12) 
plt.show()
```
